8. Fetch all students enrolled in a particular course (given course_id as input).

```python
course_id = int(input("Enter course ID: "))

# Step 1: Find all enrollments for the course
enrollments = db.enrollment.find({"course_id": course_id}, {"student_id": 1, "_id": 0})

# Step 2: Fetch matching students
student_ids = [e["student_id"] for e in enrollments]

if student_ids:
    students = db.students.find({"student_id": {"$in": student_ids}}, {"_id": 0})
    print("\nStudents enrolled in course", course_id)
    for s in students:
        print(s)
else:
    print("No students enrolled in this course.")
```

9. Write a program to display students who are not enrolled in any course.

```python
pipeline = [
    {"$lookup": {
        "from": "enrollment",          
        "localField": "student_id",   
        "foreignField": "student_id",
        "as": "enrolled"
    }},
    {"$match": {"enrolled": {"$eq": []}}},
    {"$project": {"_id": 0, "student_id": 1, "name": 1}}
]

results = list(df_s.aggregate(pipeline))
if results:
    for r in results:
        print(r)
else:
    print("All students are enrolled in at least one course.")
```

10. Perform an INNER JOIN between students and courses using Python to show student names along with the courses they are enrolled in.

```python
pipeline = [
    {"$lookup": {
        "from": "students",
        "localField": "student_id",
        "foreignField": "student_id",
        "as": "student_info"
    }},
    {"$unwind": "$student_info"},
    
    {"$lookup": {
        "from": "courses",
        "localField": "course_id",
        "foreignField": "course_id",
        "as": "course_info"
    }},
    {"$unwind": "$course_info"},
    
    {"$project": {
        "_id": 0,
        "student_name": "$student_info.name",
        "course_name": "$course_info.course_name"
    }}
]

results = list(df_e.aggregate(pipeline))
for r in results:
    print(f"{r['student_name']} → {r['course_name']}")
```

12. Add a new column email to the students table using Python and update it with random values.

```python
import numpy as np

domains = ["gmail.com", "yahoo.com", "outlook.com", "example.com"]
chars = list("abcdefghijklmnopqrstuvwxyz0123456789")

for s in df_s.find({}, {"_id": 0, "student_id": 1, "name": 1}):
    random_str = ''.join(np.random.choice(chars, 3))
    domain = np.random.choice(domains)
    school_email = f"{s['name'].lower()}{random_str}@{domain}"

    df_s.update_one(
        {"student_id": s["student_id"]},
        {"$set": {"school_email": school_email}}
    )

print("Emails added successfully!")

for s in df_s.find({}, {"_id": 0, "student_id": 1, "name": 1, "email": 1}):
    print(s)
```

13. Create a program to count how many students are enrolled per course and display results in descending order.

```python
pipeline = [
    {"$group": {"_id": "$course_name", "num_students": {"$sum": 1}}},
    {"$sort": {"num_students": -1}}
]

# Run aggregation on the collection
results = list(df_e.aggregate(pipeline))

# Print results
for doc in results:
    print(f"Course: {doc['_id']}, Students Enrolled: {doc['num_students']}")
```

14. Write a Python script that accepts a course name as input and returns all enrolled students with their grades.

```python
course_name = input("Enter course name (partial allowed): ")

# find matching courses
for course in df_c.find({"course_name": {"$regex": course_name, "$options": "i"}}):
    print(f"\nStudents enrolled in {course['course_name']}:\n")
    enrollments = df_e.find({"course_id": course["course_id"]})
    found = False
    for e in enrollments:
        student = df_s.find_one({"student_id": e["student_id"]})
        if student:
            found = True
            print("ID:", student["student_id"],
                  "| Name:", student["name"],
                  "| Grade:", student["grade"])
    if not found:
        print("No students enrolled.")
```

19. Write a program to implement transactions in Python:
○ Insert a student
○ Enroll them in a course
○ Roll back if the second query fails

```python
new_student = {"student_id": 10, "name": "John Doe", "age": 21}
new_enrollment = {"student_id": 10, "course_name": "Math", "grade": "A"}

try:
    # Start a session
    with client.start_session() as session:
        # Start a transaction
        with session.start_transaction():
            # Insert the student
            df_s.insert_one(new_student, session=session)
            
            # Simulate possible failure: uncomment next line to test rollback
            # raise Exception("Simulated failure during enrollment")
            
            # Enroll the student
            df_e.insert_one(new_enrollment, session=session)

    print("Transaction completed successfully!")

except Exception as e:
    print(f"Transaction aborted due to error: {e}")
```
